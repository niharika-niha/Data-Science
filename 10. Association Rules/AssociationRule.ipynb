```python

import pandas as pd
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# load the data
retail_data = pd.read_excel("Online retail.xlsx")
retail_data
```

```python
# count of duplicate record
retail_data.duplicated().sum()
```

```python

retail_data[retail_data.duplicated()] # to analyse duplictae variables
```

```python
retail_data=retail_data.drop_duplicates() # dropping duplicate records
retail_data.head()
```

```python
retail_data.shape
```

```python
retail_data = retail_data[ retail_data.columns[0]].str.split(',', expand=True)
retail_data.head()
```

```python
retail_data.isna().sum() # there are no missing values in the given dataset
```

```python
# replacing them with empty
retail_data_cleaned = retail_data.fillna('')
retail_data_cleaned.head()
```

```python
# performing list
transactions_data = retail_data_cleaned.apply(lambda row: [i for i in row if i != ''], axis=1).tolist()
transactions_data[:5]
```

```python
# Use TransactionEncoder to convert the transaction data into a one-hot encoded DataFrame

te = TransactionEncoder()
transactions_arr = te.fit(transactions_data).transform(transactions_data)
```

```python
# Create a DataFrame from the transformed data
t_retail_data = pd.DataFrame(transactions_arr, columns=te.columns_)
t_retail_data.head()
```

```python
# Apply the Apriori Algorithm to find frequent itemsets
# Minimum support threshold set to 0.02 (2% of transactions)
# frequent_itemsets = apriori(t_retail_data, min_support=0.1, use_colnames=True)
# frequent_itemsets

frequent_itemsets = apriori(t_retail_data, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
# Generate the Association Rules
# Using minimum confidence of 0.5 and lift greater than 1.0 for more meaningful rules

rules = association_rules(frequent_itemsets,metric="lift", min_threshold=0.7,num_itemsets=len(frequent_itemsets))
rules
```

```python
# Filter the rules with confidence >= 0.5 for stronger relationships
filtered_rules = rules[rules['confidence'] >= 0.5]
filtered_rules
```

```python
# Sort the rules by lift to identify the most interesting ones
sorted_rules = filtered_rules.sort_values(by='lift', ascending=False)
sorted_rules.head()
```

```python
# Looking for any rules with high confidence but lower lift
high_confidence_rules = sorted_rules[sorted_rules['confidence'] >= 0.8]
print("\nHigh confidence rules (confidence >= 0.8):")
print(high_confidence_rules)
```

```python
# Insights: Observing frequent patterns in the rules
print("\nSummary of the Association Rules:")
print(f"Total number of rules: {len(sorted_rules)}")
```

What is lift and why is it important in Association rules?

Lift is a measure of the strength of an association between two items in a market basket analysis, or association rule mining.

It compares the observed frequency of two items being bought together to the expected frequency if the two items were independent of each other.

formula:
Lift(X, Y) = Support(X ∩ Y) / (Support(X) * Support(Y))

Importance in Association rules:

*   Lift helps identify the strength of associations between items
*   if the lift between "avocado" and "salmon" is very high from the dataset, it suggests that customers who buy avocado are likely to buy salmon as well.

*   Lift helps eliminate unrealted items that cause misleading in associations by considering how much more likely the items are to be purchased together than by chance.



**2. What is support and Confidence. How do you calculate them?**

**Support** measures the frequency of occurrence of an itemset (a set of items) in a dataset.

*calculation:* Support(X) = (Number of transactions containing X) / (Total number of transactions)

It's calculated as the ratio of the number of transactions containing the itemset to the total number of transactions.

**Confidence** measures the reliability of an association rule.

*Calculation:*
Confidence(X -> Y) = Support(X ∪ Y) / Support(X)

It represents the probability that an itemset Y will be purchased, given that another itemset X has already been purchased.
